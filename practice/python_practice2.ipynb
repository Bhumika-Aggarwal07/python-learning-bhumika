**Q1.** Sales & Commission Analysis
 You are given the revenue (in

1. Calculate the profit for each item.
 2. Find the average profit.
 lakhs) for three electronic items:
 3. Given a commission percentage array [5, 7, 10], calculate the total commission amount for each
 item.
 4. Display all results in a formatted output using OOP (create a Product class)

```python
import numpy as np

# Step 1: Define the Product class
class Product:
    def __init__(self, names, revenue, expenses, commission_rates):
        self.names = np.array(names)
        self.revenue = np.array(revenue)
        self.expenses = np.array(expenses)
        self.commission_rates = np.array(commission_rates)
        
    def calculate_profit(self):
        self.profit = self.revenue - self.expenses
        return self.profit
    
    def average_profit(self):
        return np.mean(self.profit)
    
    def calculate_commission(self):
        # Commission = (profit × rate%) / 100
        self.commission = (self.profit * self.commission_rates) / 100
        return self.commission
    
    def display_results(self):
        print("Product Performance Report")
        print("--------------------------")
        for i in range(len(self.names)):
            print(f"Product: {self.names[i]}")
            print(f"Revenue: ₹{self.revenue[i]} Lakhs")
            print(f"Expense: ₹{self.expenses[i]} Lakhs")
            print(f"Profit: ₹{self.profit[i]:.2f} Lakhs")
            print(f"Commission: ₹{self.commission[i]:.2f} Lakhs")
            print("--------------------------")
        print(f"Average Profit: ₹{self.average_profit():.2f} Lakhs")

# Step 2: Input data
product_names = ['Laptops', 'Mobiles', 'Tablets']
revenue = [120, 100, 80]       # in lakhs
expenses = [70, 60, 50]        # in lakhs
commission_rates = [5, 7, 10]  # in %

# Step 3: Create object
products = Product(product_names, revenue, expenses, commission_rates)

# Step 4: Perform calculations
products.calculate_profit()
products.calculate_commission()

# Step 5: Display results
products.display_results()
```

 **Q3.** Student Score Evaluation
 Create NumPy arrays to store marks of 5 students in three subjects:
 maths = np.array([78, 85, 92, 68, 90])
 science = np.array([80, 83, 88, 70, 85])
 english = np.array([75, 89, 90, 72, 88])
 1. Calculate the total and average marks for each student.
 2. Find which student scored the highest average.
 3. Use OOP to create a class Student that prints the result of a student when roll number is given.

```python
import numpy as np

class Student:
    def __init__(self, roll_numbers, maths, science, english):
        self.roll_numbers = np.array(roll_numbers)
        self.maths = np.array(maths)
        self.science = np.array(science)
        self.english = np.array(english)
        
        # Calculate total and average marks
        self.total_marks = self.maths + self.science + self.english
        self.average_marks = self.total_marks / 3
        
    def show_all_results(self):
        print("Student Results Summary")
        print("-----------------------")
        for i in range(len(self.roll_numbers)):
            print(f"Roll No: {self.roll_numbers[i]}")
            print(f"Maths: {self.maths[i]}, Science: {self.science[i]}, English: {self.english[i]}")
            print(f"Total: {self.total_marks[i]}, Average: {self.average_marks[i]:.2f}")
            print("-----------------------")
        
        # Highest average
        highest_index = np.argmax(self.average_marks)
        print(f"Highest Average Scorer: Roll No {self.roll_numbers[highest_index]} with Average {self.average_marks[highest_index]:.2f}\n")
    
    def show_student_result(self, roll_no):
        if roll_no in self.roll_numbers:
            i = np.where(self.roll_numbers == roll_no)[0][0]
            print(f"Result for Roll No {roll_no}")
            print("-----------------------")
            print(f"Maths: {self.maths[i]}")
            print(f"Science: {self.science[i]}")
            print(f"English: {self.english[i]}")
            print(f"Total Marks: {self.total_marks[i]}")
            print(f"Average Marks: {self.average_marks[i]:.2f}")
        else:
            print("Invalid Roll Number!")



roll_numbers = [1, 2, 3, 4, 5]
maths = np.array([78, 85, 92, 68, 90])
science = np.array([80, 83, 88, 70, 85])
english = np.array([75, 89, 90, 72, 88])

students = Student(roll_numbers, maths, science, english)

# Display all results
students.show_all_results()

# Show result for a specific student (for example Roll No 3)
students.show_student_result(3)

        
```

**Q4**  Medicine Production Analysis
 A pharmaceutical company produces 3 types of medicines.
 The data for production cost, selling price, and units sold are:
 cost = np.array([12, 8, 10])
 price = np.array([20, 15, 18])
 sold = np.array([5000, 8000, 6000])
 1. Calculate profit per medicine.
 2. Calculate total profit for all medicines.
 3. Compute the mean, median, and standard deviation of profits.

```python
cost = np.array([12, 8, 10])
price = np.array([20, 15, 18])
sold = np.array([5000, 8000, 6000])

profit = (price - cost)*sold
print(f" profit per medicine : {profit}")

total_profit = np.sum(profit)
print(f"total profit for all medicines : {total_profit}")

print(np.mean(profit))
print(np.median(profit))
print(np.std(profit))
```

**Q8.** OOP + NumPy Integration
 Create a Product class with attributes name, cost, and price.
 Use NumPy arrays to store multiple products and:
 1. Compute profit percentage for each.
 2. Display product with maximum profit

```python
class Product:
    def __init__(self , name , cost , price):
        self.name = np.array(name)
        self.cost = np.array(cost)
        self.price = np.array(price)

    def profit_percentage(self):
        self.profit = ((self.price - self.cost ) / self.cost) * 100
        return (f" profit percentage for each product is {self.profit}")

p1 = Product(['a','b','c'],[12,34,98],[34,57,120])
print(p1.profit_percentage())
max_p_index = np.argmax(p1.profit)
print(p1.name[max_p_index])

    
```

**Q9**. Employee Salary Adjustment
 An array stores base salaries of employees:
 salary = np.array([45000, 52000, 60000, 48000, 75000])
 1. Increase salary by 10% for those earning below 50,000.
 2. Calculate the new mean salary.
 3. Find the difference between highest and lowest salary after increment

```python
salary = np.array([45000, 52000, 60000, 48000, 75000] , dtype = 'float64')

salary_filter = salary < 50_000
salary[salary_filter] *= 1.1
print(salary)
print(np.mean(salary))
print(np.max(salary) - np.min(salary))
```

**Q 10**  An investor invests in 4 companies:
 investment = np.array([100000, 150000, 120000, 180000])
 returns = np.array([110000, 160000, 130000, 200000])
 1. Calculate the return percentage for each.
 2. Find the average return percentage.
 3. Use NumPy logical operations to identify which companies gave >10% return

```python
companies = np.array(['a' , 'b', 'c', 'd'])
investment = np.array([100000, 150000, 120000, 180000]) 
returns = np.array([110000, 160000, 130000, 200000])

return_percentage =(( returns - investment) / investment) * 100
print(return_percentage)

avg_return_percentage =  np.mean(return_percentage)
high_return_mask = return_percentage > 10
print(companies[high_return_mask])
```
