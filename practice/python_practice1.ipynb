**Q1** Given a string, find the longest consecutive substring containing only alphabets.
You can assume that the string will have at least one alphabet.
Note: If there are two such substrings then you should print the substring starting at a lower index. (For example: in the string "here123were23we", you should print "here".


```python
string=input()
substrings_list = []
substring = []
for i in string :
    if i.isalpha():
        substring.append(i)
    else:
        if substring:
            word = ''.join(substring)
            substrings_list.append(word)
            substring = []
            
if substring:
    word = ''.join(substring)
    substrings_list.append(word)
            
max_len = 0
longest_substring = ''
for j in substrings_list:
    if len(j) > max_len:
        longest_substring = j
        max_len = len(j)

print(longest_substring)
```

**Q2**  Given a positive integer n, your task is to print sum of all prime numbers up to the number n, starting from 2.

```python
n = int(input())  # print all prime numbers till n
sum_total = 0
for i in range(2, n + 1):
    for j in range(2, int(i ** 0.5) + 1):
        if i % j == 0:
            break
    else:
        sum_total += i

print(sum_total)
        
```

**Q6** Simon runs a pharmaceutical factory. Only customers who have been pre-approved can enter the manufacturing facility and do a quality check for their order.
The pre-approved customers get a customer ID which needs to be validated at entrance. 
A valid customer ID:
- It will have groups of three characters separated by "_".
- Can only contain numbers 0-9 and characters a-z and A-Z.(other than "_")
- Contains 12 characters excluding the "_"
- A character cannot be repeated more than thrice.

You have to write Python code to check for valid customer IDs.
The input will have a customer ID and the output should say "valid" if the entered ID is valid otherwise it should say "invalid".

```python
cid = input()
is_grouped_3 = True
ch_check = False
no_ch_check = True
ch_repeat = True

words_lst = cid.split('_')
for word in words_lst :
    if len(word) != 3 :
        is_grouped_3 = False
        break

chars = ''.join(words_lst)
if len(chars) != 12:
    no_ch_check = False

for i in chars:
    if i.isalnum():
        ch_check = True
    else:
        ch_check = False
        break

count_dict ={}
for i in chars:
    count_dict[i] = count_dict.get(i , 0) + 1

for i in count_dict.values() :
    if i > 3:
        ch_repeat = False
        break

if is_grouped_3 and ch_check and no_ch_check and ch_repeat :
    print('valid')
else:
    print('invalid')
        
```

**Q10.** An online shopping platform manages various products.
All products have a name and price, but electronics include a warranty period, while clothes include a size.
Premium customers get an extra 10% discount on all purchases.

Design a Python program to handle billing for both normal and premium customers using inheritance and method overriding.

```python
def premium_price(price):
    discounted_price = 0.9 * price
    return discounted_price

class Products:
    def __init__(self, name , price):
        self.name = name
        self.price = price

    def get_final_price(self):
        return premium_price(self.price)

class Electronics(Products) :
    def __init__(self, name , price , warranty):
        super().__init__(name , price)
        self.warranty = warranty

    def show_details(self):
        is_premium = input("Are you premium customer (True / False )")
        final_price = self.get_final_price()
        return (f"{self.name} costs {final_price} has {self.warranty} years warranty.")

class Clothes(Products) :
    def __init__(self, name , price , size):
        super().__init__(name , price)
        self.size = size

    def show_details(self):
        is_premium = input("Are you premium customer (True / False )")
        final_price = self.get_final_price()
        return (f"{self.name} costs {final_price} has {self.size} size.")

mobile = Electronics('M51' , 23000 , 1)
print(mobile.show_details())

top = Clothes( 'Top' , 300 , 'M')
print(top.show_details())
```

**Q 11** bank maintains multiple account types.
Each customer has a name and account balance.
A savings account earns interest, while a current account allows an overdraft limit.
The bank also has premium customers who can hold both types of accounts and get extra benefits.

Use hybrid inheritance to model this system and demonstrate how each account type performs its operations differently.

```python
class Accounts:
    def __init__(self, name, account_balance):
        self.name = name
        self.account_balance = account_balance


class Savings_Acc(Accounts):
    def __init__(self, name, account_balance, rate_of_interest, time):
        # Call base class explicitly
        Accounts.__init__(self, name, account_balance)
        self.rate_of_interest = rate_of_interest
        self.time = time

    def earned_interest(self):
        # Simple Interest formula
        interest = (self.account_balance * self.rate_of_interest * self.time) / 100
        return interest

    def show_details(self):
        return (
            f"Savings Account of {self.name}:\n"
            f"Balance: ₹{self.account_balance}\n"
            f"Interest earned: ₹{self.earned_interest():.2f}\n"
        )


class Overdraft_Acc(Accounts):
    def __init__(self, name, account_balance, overdraft_amt):
        # Call base class explicitly
        Accounts.__init__(self, name, account_balance)
        self.overdraft_amt = overdraft_amt

    def show_details(self):
        return (
            f"Current Account of {self.name}:\n"
            f"Balance: ₹{self.account_balance}\n"
            f"Overdraft Limit: ₹{self.overdraft_amt}\n"
        )


class PremiumCustomer(Savings_Acc, Overdraft_Acc):
    def __init__(self, name, account_balance, rate_of_interest, time, overdraft_amt, cashback):
        # Explicitly call both parents
        Savings_Acc.__init__(self, name, account_balance, rate_of_interest, time)
        Overdraft_Acc.__init__(self, name, account_balance, overdraft_amt)
        self.cashback = cashback

    def show_details(self):
        total_balance = self.account_balance + self.earned_interest() + self.cashback
        return (
            f"Premium Customer: {self.name}\n"
            f"Base Balance: ₹{self.account_balance}\n"
            f"Interest Earned: ₹{self.earned_interest():.2f}\n"
            f"Cashback: ₹{self.cashback}\n"
            f"Overdraft Limit: ₹{self.overdraft_amt}\n"
            f"Total Balance (with cashback & interest): ₹{total_balance:.2f}\n"
        )


# Demonstration
s1 = Savings_Acc('ACXY890', 20000, 2, 2)
print(s1.show_details())

s2 = Overdraft_Acc('ACFDHUIDHL', 40000, 5000)
print(s2.show_details())

p1 = PremiumCustomer('BHUMIKA', 50000, 3, 2, 10000, 500)
print(p1.show_details())
```

**Q12** At a university, people can be either students, faculty members, or researchers.
A teaching assistant is both a student and a researcher who also helps in teaching.

Write a Python program to represent this situation using multiple inheritance.
Demonstrate how a teaching assistant can access properties from both parent roles, and explain the order of constructor calls.

```python
class Students:
    def __init__(self, name, program):
        self.name = name
        self.program = program

    def show_details(self):
        return f"{self.name} is a student enrolled in {self.program}"


class Faculty:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def show_details(self):
        return f"{self.name} is a faculty member, teaching {self.subject}"


class Researchers:
    def __init__(self, name, topic):
        self.name = name
        self.topic = topic

    def show_details(self):
        return f"{self.name} is a researcher, researching on {self.topic}"


# Teaching Assistant inherits from BOTH Students and Researchers
class Teaching_Asst(Students, Researchers):
    def __init__(self, name, program, topic):
        # explicitly call both parent constructors
        Students.__init__(self, name, program)
        Researchers.__init__(self, name, topic)

    def show_details(self):
        return (
            f"{self.name} is a Teaching Assistant pursuing {self.program} "
            f"and researching on {self.topic}."
        )


# Demonstration
s1 = Students('Arush', 'CSE')
print(s1.show_details())

f1 = Faculty('Puneet', 'Physics')
print(f1.show_details())

r1 = Researchers('Priya', 'Atomic Studies')
print(r1.show_details())

t1 = Teaching_Asst('Urvi', 'CSE', 'Quantum Computing')
print(t1.show_details())
```

You must design a system that allows:

Displaying complete ride details.

Comparing the fares between any two rides and printing which one costs more per trip.
(Hint: Use inheritance and method overriding for different vehicle types.)

```python
class Vehicles:
    def __init__(self, brand , max_speed):
        self.brand = brand
        self.max_speed = max_speed

class Cars(Vehicles):
    def __init__(self, brand , max_speed , trip_fare):
        super().__init__(brand , max_speed )
        self.trip_fare = trip_fare

    def show_details (self):
        return (f"This is {self.brand} having maximum speed as {self.max_speed} and trip fare as {self.trip_fare}")

    def compare_fare(self, other):
        if self.trip_fare > other.trip_fare:
            diff = self.trip_fare - other.trip_fare
            return f"{self.brand} (Car) costs ₹{diff} more per trip than {other.brand}."
        elif self.trip_fare < other.trip_fare:
            diff = other.trip_fare - self.trip_fare
            return f"{other.brand} costs ₹{diff} more per trip than {self.brand} (Car)."
        else:
            return f"Both {self.brand} and {other.brand} have the same trip fare."
            
class Bikes(Vehicles):
    def __init__(self, brand , max_speed , trip_fare):
        super().__init__(brand , max_speed)
        self.trip_fare = trip_fare
    def show_details (self):
        return (f"This is {self.brand} having maximum speed as {self.max_speed} and trip fare as {self.trip_fare}")

    def compare_fare(self, other):
        if self.trip_fare > other.trip_fare:
            diff = self.trip_fare - other.trip_fare
            return f"{self.brand} (Bike) costs ₹{diff} more per trip than {other.brand}."
        elif self.trip_fare < other.trip_fare:
            diff = other.trip_fare - self.trip_fare
            return f"{other.brand} costs ₹{diff} more per trip than {self.brand} (Bike)."
        else:
            return f"Both {self.brand} and {other.brand} have the same trip fare."



car1 = Cars('Mahendras','180 km/hr' , 500)
print(car1.show_details())

bike1 = Bikes('Hero','180 km/hr' , 300)
print(bike1.show_details())

print(car1.compare_fare(bike1))
```

**Q14** A hospital wants an automated billing system. Each doctor specializes in a field and charges a consultation fee. Some doctors also perform surgeries that add extra charges. When a patient visits, the total bill depends on consultation + surgery + hospital service fee. Write a program that can display each doctor’s total earnings for multiple patients and show which doctor earned the most.

```python
class Doctor:
    def __init__(self, name, specialization, consultation_fee):
        self.name = name
        self.specialization = specialization
        self.consultation_fee = consultation_fee
        self.total_earnings = 0  # total earnings from all patients

    def calculate_bill(self, hospital_service_fee):
        # consultation + service fee
        total = self.consultation_fee + hospital_service_fee
        self.total_earnings += total
        return total

    def show_details(self):
        return f"Dr. {self.name} ({self.specialization}) - Total Earnings: ₹{self.total_earnings}"


class Surgeon(Doctor):
    def __init__(self, name, specialization, consultation_fee, surgery_fee):
        super().__init__(name, specialization, consultation_fee)
        self.surgery_fee = surgery_fee

    def calculate_bill(self, hospital_service_fee):
        # consultation + surgery + service fee
        total = self.consultation_fee + self.surgery_fee + hospital_service_fee
        self.total_earnings += total
        return total

    def show_details(self):
        return (f"Dr. {self.name} ({self.specialization}, Surgeon) - "
                f"Total Earnings: ₹{self.total_earnings}")


hospital_service_fee = 500  # constant for all patients

# Create doctor objects
d1 = Doctor("Mehta", "Dermatologist", 800)
d2 = Surgeon("Verma", "Neurosurgeon", 1500, 5000)
d3 = Surgeon("Sharma", "Cardiac Surgeon", 1200, 7000)

# Patient visits (each call simulates a patient visit)
print(f"Bill (Patient 1): ₹{d1.calculate_bill(hospital_service_fee)}")
print(f"Bill (Patient 2): ₹{d2.calculate_bill(hospital_service_fee)}")
print(f"Bill (Patient 3): ₹{d3.calculate_bill(hospital_service_fee)}")
print(f"Bill (Patient 4): ₹{d2.calculate_bill(hospital_service_fee)}")  # Verma sees one more patient

# Display total earnings
print("\n--- Doctor Earnings ---")
print(d1.show_details())
print(d2.show_details())
print(d3.show_details())
```
