**Q 1**Employee Salary System

A company wants to automate salary computation.
Every employee has a base salary.
Managers get an additional bonus (10% of their base salary), while Developers get project allowances (₹500 per project).

Write a Python program using inheritance that:

Defines classes for Employee, Manager, and Developer.

Calculates and displays the total salary for both types.

Displays which employee earns more.

```python
class Employee :
    def __init__(self, name , base_salary):
        self.name = name
        self.base_salary = base_salary 

    def get_salary(self):
        return (f"{self.name} has {self.base_salary} salary")

class Manager(Employee):
    def __init__(sself ,name , base_salary ):
        super().__init__(name , base_salary)

    def get_salary(self):
        total_salary = self.base_salary * 1.1

        return (f"{self.name} has {total_salary} salary")


class Developer(Employee):
    def __init__(self,name , base_salary,n):
        super().__init__(name , base_salary)
        self.n = n

    def get_salary(self):
        total_salary = self.base_salary + (self.n * 500)

        return (f"{self.name} has {total_salary} salary")


e1 = Employee('a' ,1200)
m1 = Manager('b' , 1300)
d1 = Developer('c' , 1200 ,6)

print(e1.get_salary())
print(m1.get_salary())
print(d1.get_salary())
```

**Q2** At a university, there are Students and Faculty members.
A Teaching Assistant (TA) is both a Student and also assists a Faculty member in teaching.

Write a Python program that:

Stores basic details for Student and Faculty.

Allows a Teaching Assistant to access information from both.

Displays the TA’s full role details clearly.

```python
class Faculty :
    def __init__(self , name , subject):
        self.name = name
        self.subject = subject

    def show_details(self):
        return (f"{self.name} is a faculty member who teaches {self.subject}")

class Student :
    def __init__(self ,name , program):
        self.name = name
        self.program = program

    def show_details(self):
        return (f"{self.name} is a student enrolled in {self.program}")


class Teaching_asst(Faculty ,  Student):
    def __init__(self ,name , subject , program):
        Faculty.__init__(self ,name , subject)
        Student.__init__(self ,name , program)

    def show_details(self):
        return (f"{self.name} is a teaching assisstant who teaches {self.subject} and is enrolled in {self.program}")



f1 = Faculty('a' , 'MTH165')
s1 = Student('b' , 'CSE')
t1 = Teaching_asst('c' , 'MTH165' , 'CSE')

print(f1.show_details())
print(s1.show_details())
print(t1.show_details())
```

```python
def solution(array):
    n = len(array)
    mid = n// 2
    min_diff = max(array)

    for d in range(1, mid + 1):
        left = array[mid - d]
        right = array[mid + d]
        diff = abs(left - right)
        if diff < min_diff:
            min_diff = diff

    return min_diff


print(solution([5,9,7,2,4,5,9]))
```

###### Double pairs are pair of 2 numbers, such that one of them is double of the other. You are given an array, wherein all integers are unique. You have to find the number of such pairs.Return the number of such pairs

```python
def find_pairs(nums):
    count = 0
    for i in nums:
        if 2*i in nums :
            print(f"{i} , {2*i}")
            count += 1
            
    return count
find_pairs([5,6,7,12,14,28])
```

```python
# Make the number max

def beauty_num(num):
    digits = list(str(num))
    n = len(digits)

    last = [-1] * 10
    for i, d in enumerate(digits):
        last[int(d)] = i

    for i in range(n):
        current_digit = int(digits[i])
        for d in range(9, current_digit, -1):
            if last[d] > i:
                swap_idx = last[d]
                digits[i], digits[swap_idx] = digits[swap_idx], digits[i]
                return int("".join(digits))

    return num

beauty_num(2736)
	
```
